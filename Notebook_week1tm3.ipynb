```python
df_step = pd.read_csv("steps.csv", sep=";")
df_survey = pd.read_csv('survey.csv')

df = df_survey.merge(df_step, on = "id")
df.head(10)
```

## Predictive model

Met liniare regressie kun je een liniair verband tussen twee variabelen vaststellen. 

```python
x = df[['steps_mean']]
y = df[['height', 'weight', 'age', 'education']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=23)
```

```python
lm = LinearRegression() 

lm = lm.fit(x_train, y_train)
```

## Evaluation

```python
y_test_p = lm.predict(x_test)
r2_score(y_test, y_test_p)
```
